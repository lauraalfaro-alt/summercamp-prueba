```python
# Importamos las herramientas que usaremos hoy
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Creamos una conexión a un archivo de base de datos virtual en Colab
conexion = sqlite3.connect("torneo_smash.db")

# 2. Leemos nuestro archivo CSV usando Pandas
df_original = pd.read_csv("smash_dataset_clean.csv", encoding="latin1")

# 3. ¡La Magia! Guardamos toda esa información dentro de una tabla SQL llamada 'combates'
df_original.to_sql("combates", conexion, if_exists="replace", index=False)

print("✅ ¡Éxito total! La base de datos relacional 'torneo_smash.db' ha sido creada y la tabla 'combates' está lista.")
```

```python
consulta_sql = """
SELECT nombre, personaje, edad
FROM combates
Order By edad ASC
"""

df_filtrado = pd.read_sql_query(consulta_sql, conexion)

print("📊 Datos extraídos mediante SQL con éxito:")
df_filtrado.head(24)



```

```python
from pandas.core.arrays import integer
import matplotlib.ticker as ticker
sns.set_theme(style="darkgrid")

plt.figure(figsize=(14,7))

grafico_barras = sns.barplot(
    data=df_filtrado,
    x="Nombre",
    y="Edad",
    palette="coolwarm",
    errorbar=None
)

plt.title("Edad de los participantes", fontsize=18, fontweight="bold", pad=15)
plt.xlabel("Nombre de los jugadores", fontsize=14, labelpad=12)
plt.ylabel("Edad de los jugadores", fontsize=14, labelpad=12)

plt.xticks(rotation=45)

plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.show()

```

```python
# Importamos las herramientas que usaremos hoy
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Creamos una conexión a un archivo de base de datos virtual en Colab
conexion = sqlite3.connect("torneo_smash.db")

# 2. Leemos nuestro archivo CSV usando Pandas
df_original = pd.read_csv("smash_dataset_clean.csv", encoding="latin1")

# 3. ¡La Magia! Guardamos toda esa información dentro de una tabla SQL llamada 'combates'
df_original.to_sql("combates", conexion, if_exists="replace", index=False)

print("✅ ¡Éxito total! La base de datos relacional 'torneo_smash.db' ha sido creada y la tabla 'combates' está lista.")
```

```python
consulta_sql = """
SELECT nombre, personaje, victorias
FROM combates
Order By victorias ASC
"""

df_filtrado = pd.read_sql_query(consulta_sql, conexion)

print("📊 Datos extraídos mediante SQL con éxito:")
df_filtrado.head(24)
```

```python
from pandas.core.arrays import integer
import matplotlib.ticker as ticker
sns.set_theme(style="darkgrid")

plt.figure(figsize=(14,7))

grafico_barras = sns.barplot(
    data=df_filtrado,
    x="Nombre",
    y="Victorias",
    palette="coolwarm",
    estimator='sum',
    errorbar=None
)

plt.title("Victorias de los participantes", fontsize=18, fontweight="bold", pad=15)
plt.xlabel("Nombre de los jugadores", fontsize=14, labelpad=12)
plt.ylabel("Victorias de los jugadores", fontsize=14, labelpad=12)

plt.xticks(rotation=45)

plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.show()
```
